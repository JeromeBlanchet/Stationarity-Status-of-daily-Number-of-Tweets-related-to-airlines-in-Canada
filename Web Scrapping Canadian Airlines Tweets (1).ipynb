```python
def __TWEETQUERY__(LIST):

    start_time = time.clock()
    
    dataset = pd.DataFrame(columns=['Tweet_Date',
                                'Tweet_Author',
                                'Tweet_Retweets',
                                'Tweet_Tweet'])

    for i in LIST:

        text_query = 'airline canada'
        since_date = i[0]
        until_date = i[1]
        count = 0

        tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setSince(since_date)\
        .setUntil(until_date).setMaxTweets(count)
        tweets = got.manager.TweetManager.getTweets(tweetCriteria)
        text_tweets = [[tweet.date, tweet.username, tweet.retweets, tweet.text] for tweet in tweets]
    
        for y in text_tweets:
            dataset = dataset.append({'Tweet_Date': y[0],
                                      'Tweet_Author': y[1],
                                      'Tweet_Retweets': y[2],
                                      'Tweet_Tweet': y[3]}, ignore_index=True)
        
        end_time = time.clock()

        print("")
        print("Total Estimation Running Time:")
        print(end_time - start_time, "Seconds")
        
    return dataset
```

```python
window_size = 365
total_range_ = range(0, weekly1.shape[0], 1)
total_range_len = len(total_range_)
used_range = total_range_len - window_size
xxxx = range(0, used_range, 1)
list_of_interval = [[1+i, window_size+i] for i in xxxx]

#Setting up ADF test function
from statsmodels.tsa.stattools import adfuller

adf_test_stat = []
adf_test_crit1 = []
adf_test_crit2 = []
adf_test_crit3 = []
adf_test_pvalue = []
adf_test_nblag = []
adf_test_nbobs = []
adf_test_bestic = []

for interval in list_of_interval:
    result = adfuller(weekly1[interval[0]:interval[1]].values, autolag='BIC', regression='c') #{‘c’,’ct’,’ctt’,’nc’}
    adf_test_stat.append(result[0])
    adf_test_crit1.append(list(result[4].values())[0])
    adf_test_crit2.append(list(result[4].values())[1])
    adf_test_crit3.append(list(result[4].values())[2])
    adf_test_pvalue.append(result[1])
    adf_test_nblag.append(result[2])
    adf_test_nbobs.append(result[3])
    adf_test_bestic.append(result[5])    

adf_test_nblag_clean = adf_test_nblag
       
for index, item in enumerate(adf_test_nblag):
    if item == 0:
        adf_test_nblag_clean[index] = None    
        
fig, ax = plt.subplots(figsize=FIGSIZE)

plt.plot(adf_test_stat, 'r')
plt.plot(adf_test_crit1, 'b')
plt.plot(adf_test_crit2, 'b')
plt.plot(adf_test_crit3, 'b')
plt.plot(adf_test_pvalue, 'g')
plt.plot(adf_test_nblag, 'k')

plt.title("Evolution of the Augmented Dickey-Fuller Test Over Time\n\
\n\
ADF Test Based on Daily Number of Tweet including words 'Airline' & 'Canada'\n\
January 2007 - April 2020, 4,745 time step, 138,222 Individual Tweet\nNo Seasonal Adjustment\n\
\n\
365 days (1 year) Rolling Window Regression\n\
With Constant (No Linear or Quadratic Trend)\n\
Max Number of Lags of Short-Run Variables (Delta y) Based on Best BIC Criterion\n\
\n\
Red = ADF Statistical Value\n\
Blue = Critical Value (1% (Lowest), 5% & 10% (Highest))\n\
Green = P-Value\n\
Black = Max Number of Lags of Short-Run Variable (Delta y)(None if nb of Lags is 0) \n\
----------------------------------------------------------------------------------------------\n\
------------------------------------------------------------------------------------------------------------------------", 
fontdict=FONT, y=1.2)

plt.xlabel("Rolling Window ADF Regression Number\n\
Last Rolling Window Regression is [April 2019, April 2020]", fontdict=FONT)

plt.ylabel("Statistical & Critical Value", fontdict=FONT)

plt.ylim((-22,12))
ax.set_frame_on(False)
plt.grid(True)
plt.show()
```
